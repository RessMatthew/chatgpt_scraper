```python
import re
import numpy as np
import copy
import pandas as pd
```

```python
# 文件路径
file_path = '../datasets/our_data.csv'

# 读取数据，跳过前两行，手动指定列名
data = pd.read_csv(
    file_path,
    skiprows=2,  # 跳过前两行
    header=None,  # 不使用文件中的列名
    names=["scenario", "type", "description", "data", "definition", "output", "answer"],  # 指定列名
    usecols=[0, 1, 2, 3, 4, 5, 6]  # 只读取实际需要的列
)

# 显示前几行数据，验证是否正确对齐
data.head()
```

```python
# 打乱数据顺序
shuffled_data = data.sample(frac=1, random_state=123).reset_index(drop=True)

# 保存打乱后的数据到新的文件
shuffled_file_path = './datasets/fine_data.csv'  # 替换为保存文件的实际路径
shuffled_data.to_csv(shuffled_file_path, index=False)

# 返回新的文件路径
print(f"打乱后的数据已保存至：{shuffled_file_path}")
```
